```python
import random
import numpy as np

p = {'Content' : 0.3,
     'Moyen content' : 0.4,
     'Pas content' : 0.1,
'Très content' : 0.2} 

def opinion_utilisateur(p):
    u = np.random.random()
    s = 0
    for k in p.keys():
        s += p[k]
        if u < s:
            return k
        
print(opinion_utilisateur(p))

def opinion(n,p):
    z= {'Content' : 0,
     'Moyen content' : 0,
     'Pas content' : 0,
     'Très content' : 0}
    for i in range(0,n):
        z[opinion_utilisateur(p)]+=1
    return z

def opinion_utilisateur_suivant(D,p,pC,pTC,pMC,pPC):
    for i in D:
        if D[i]=='Content':
            D[i]= opinion_utilisateur(pC)
        if D[i]=='Très content':
            D[i]=opinion_utilisateur(pTC)
        if D[i]=='Moyen content':
            D[i]= opinion_utilisateur(pMC)
        if D[i]=='Pas content':
            D[i]=opinion_utilisateur(pPC)
    return D
```

```python
#GRAPHIQUE MODELISANT L'OPINION DE CHAQUE PERSONNE DE LA COMMMUNAUTE POUR UN SEUL JEU
pi = 500
#def opinion_depart(pi):
import matplotlib.pyplot as plt
import networkx as nx
import random


G = nx.random_geometric_graph(pi, 0.125)
pos = nx.get_node_attributes(G, 'pos')
D = dict()
E = {'Très content' : "green",
         'Content' : "blue", 
         'Moyen content': "orange", 
         'Pas content' : "red"
        }
P = dict()
for i in range(0,pi) :
    pos[i] = random.uniform(-0.025, 1.025), random.uniform(-0.025, 1.025)
    D[i] = opinion_utilisateur(p)
    P[i] = pos[i]
plt.figure(figsize=(8, 8))

nx.draw_networkx_nodes(G, pos,
                        nodelist = [i for i in range(0,pi)],
                        node_size=20,
                        node_color= [E[D[i]] for i in range(0,pi)])

    #nx.draw_networkx_edges(G, pos, nodelist = [i for i in range(0, 500)])
    #nx.draw_networkx_edges(G, pos, nodelist = [i for i in range(500,1000)])
plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis('off')
plt.show()
```

```python
pC={'Content' : 0.55,'Moyen content' : 0.15,'Pas content' : 0,'Très content' : 0.3}
pMC={'Content' : 0.5,'Moyen content' : 0.3,'Pas content' : 0.2,'Très content' : 0}
pTC={'Content' : 0.2,'Moyen content' : 0,'Pas content' : 0,'Très content' : 0.8}
pPC={'Content' : 0,'Moyen content' : 0.05,'Pas content' : 0.95,'Très content' : 0}

D = opinion_utilisateur_suivant(D,p,pC,pTC,pMC,pPC)

nx.draw_networkx_nodes(G,P,nodelist = [i for i in range(0,500)], node_size = 20, node_color = [E[D[i]] for i in range(0,500)])

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis('off')
plt.show()
```
